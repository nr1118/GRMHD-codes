##  Second order Runge-Kutta convergence testing using the L2 norm. I used dt = dx/2. I took the values of solution at 3 times, namely 0,pi, and two pi.

```python
c = 1.0
v = 0.5
```

```python
x = np.linspace(0,2*pi,10)
t = np.array([0,np.pi,2*np.pi])
U = np.array([[0,0.642788,0.984808,0.866025,0.34202,-0.34202,-0.866025,-0.984808,-0.642788,0],
              [-0.0520881,-0.195464,-0.24738,-0.183544,-0.0338262,0.13172,0.235632,0.22929,0.11566,-0.0520881],
              [0.0252118,0.0552173,0.0593859,0.0357673,-0.00458732,-0.0427954,-0.0609791,-0.05063,-0.0165905,0.0252118]])
              
```

```python
dx = x[1]-x[0]

cot = np.cos(x)/np.sin(x)
predicted_value_t0 = np.exp(-1*(c*cot+v)*t[0])*np.sin(x)

predicted_value_t1 = np.exp(-1*(c*cot+v)*t[1])*np.sin(x)

predicted_value_t2 = np.exp(-1*(c*cot+v)*t[2])*np.sin(x)

```

```python
fig, ax = pyplot.subplots(1,3, figsize=(12, 5))
ax[0].set_title("t_0= "+str(t[0])+' s')
ax[0].plot(x,U[0])
ax[0].plot(x,predicted_value_t0)

ax[1].set_title("t_0= "+str(t[1])+' s')
ax[1].plot(x,U[1])
ax[1].plot(x,predicted_value_t1)

ax[2].set_title("t_0= "+str(t[2])+' s')
ax[2].plot(x,U[2])
ax[2].plot(x,predicted_value_t2)
```

## $$L_{2} = \sqrt{\sum_{i}^{n} (y_{i}-f(x_{i}))^{2}}$$

```python
l2_t0 = np.sqrt(np.sum(np.power((U[0]-predicted_value_t0),2)))
print(l2_t0) 
```

```python
l2_t1 = np.sqrt(np.sum(np.power((U[1]-predicted_value_t1),2)))
print(l2_t1)
```

```python
l2_t2 = np.sqrt(np.sum(np.power((U[2]-predicted_value_t2),2)))
print(l2_t2)
```
