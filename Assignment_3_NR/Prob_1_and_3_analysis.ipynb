##  Center differencing convergence testing using the L2 norm. I used dt = dx/2. I took the values of solution at 3 times, namely 0,pi, and two pi.

```python
x = np.linspace(0,2*pi,10)
t = np.array([0,np.pi,2*np.pi])
U = np.array([[0,0.642788,0.984808,0.866025,0.34202,-0.34202,-0.866025,-0.984808,-0.642788,0],
              [-0.523267,-1.34302,-1.53436,-1.00775,-0.00960911,0.99303,1.53102,1.35263,0.541327,-0.523267],
              [1.53397,2.38009,2.11254,0.856506,-0.800293,-2.08263,-2.39047,-1.57979,-0.0299099,1.53397]])
```

```python
dx = x[1]-x[0]

predicted_value_t0 = np.sin(x-t[0])

predicted_value_t1 = np.sin(x-t[1])

predicted_value_t2 = np.sin(x-t[2])
```

```python
fig, ax = pyplot.subplots(1,3, figsize=(12, 5))
ax[0].set_title("t_0= "+str(t[0])+' s')
ax[0].plot(x,U[0])
ax[0].plot(x,predicted_value_t0)

ax[1].set_title("t_0= "+str(t[1])+' s')
ax[1].plot(x,U[1])
ax[1].plot(x,predicted_value_t1)

ax[2].set_title("t_0= "+str(t[2])+' s')
ax[2].plot(x,U[2])
ax[2].plot(x,predicted_value_t2)
```

## $$L_{2} = \sqrt{\sum_{i}^{n} (y_{i}-f(x_{i}))^{2}}$$

```python
l2_t0 = np.sqrt(np.sum(np.power((U[0]-predicted_value_t0),2)))
print(l2_t0) 
```

```python
l2_t1 = np.sqrt(np.sum(np.power((U[1]-predicted_value_t1),2)))
print(l2_t1)
```

```python
l2_t2 = np.sqrt(np.sum(np.power((U[2]-predicted_value_t2),2)))
print(l2_t2)
```
